```python
import os
import zipfile

# 1. Let's see if your archive file actually has data in it
if not os.path.exists('archive.zip'):
    print("❌ Error: 'archive.zip' does not exist in this folder at all!")
elif os.path.getsize('archive.zip') == 0:
    print("❌ Error: 'archive.zip' is empty (0 bytes). Your download failed or was interrupted.")
else:
    print(f"✓ Found 'archive.zip' ({os.path.getsize('archive.zip') / (1024*1024):.2f} MB)")

    # 2. Extract using a clean, safe local path relative to your notebook
    target_folder = "./extracted_animals"
    os.makedirs(target_folder, exist_ok=True)
    
    print("Extracting files now...")
    with zipfile.ZipFile('archive.zip', 'r') as zip_ref:
        zip_ref.extractall(target_folder)
    print("✓ Extraction process finished!")

    # 3. Let's immediately prove if files exist by scanning the new folder
    contents = os.listdir(target_folder)
    print(f"\nWhat is actually inside '{target_folder}' right now?")
    print(contents)
    
    # Check deeper if something is there
    if len(contents) > 0:
        subfolder = os.path.join(target_folder, contents[0])
        if os.path.isdir(subfolder):
            print(f"Inside subfolder '{contents[0]}':", os.listdir(subfolder)[:5])
```
