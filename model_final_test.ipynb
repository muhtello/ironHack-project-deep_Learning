# Final Test - Winner Only

Run this **once**, after all model variants are compared by validation
metrics in `model_tracking.csv`. It loads the single best model and scores
it on the held-out **test** set (untouched during model selection) to report
one honest, real-world number for the README.

Prereq: the winning notebook must save its model (e.g. `model.save(...)`).

```python
import os
import sys

sys.path.append(os.path.join(os.getcwd(), "preprocessing", "label_mapping"))
sys.path.append(os.path.join(os.getcwd(), "preprocessing", "data_loader"))
sys.path.append(os.path.join(os.getcwd(), "evaluation", "model_metrics"))

from tensorflow import keras

from label_mapping import build_labeled_dataset
from data_loader import build_train_val_test_generators
from model_metrics import debug_model, evaluate_model
```

```python
# Same seed=42 default -> identical split, so this is the SAME held-out
# test set the models never saw during selection.
df = build_labeled_dataset()
train_generator, val_generator, test_generator = build_train_val_test_generators(
    df, project_root=os.getcwd(), image_size=(128, 128)
)
```

```python
# Overall best = transfer_mobilenet (val 0.940), the model shipped in the demo.
# Scored here on the held-out test set. (gap_cnn was the best FROM-SCRATCH
# model at test 0.739; transfer learning is the bonus that beat it
WINNER_PATH = "models_saved/transfer_mobilenet.keras"
best_model = keras.models.load_model(WINNER_PATH)
best_model.summary()
```

```python
# One-time test-set evaluation: per-class report + confusion matrix...
debug_model(best_model, test_generator)
```

```python
# ...and the headline test scores to quote in the README (NOT written to the
# per-model tracking sheet - that one stays validation-only).
evaluate_model(best_model, test_generator, "WINNER (test set)")
```
